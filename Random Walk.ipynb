```
import math
import random
import matplotlib.pyplot as plt


# number of times to run this experiment
n = 10000

# number of iterations in each case, i.e. number of steps in walk
t = 1000 



distance = []


for j in range(n):
    l = []
    
    x = 0
    y = 0

    
    if j%1000==0:
      print(j)

    for i in range(t):
        a = random.randint(0,359)

        x += math.cos((math.pi * a) / 180)
        y += math.sin((math.pi * a) / 180)

        d = (((x)**2)+((y)**2))**0.5
        l.append(d)
    distance.append(l)
```

```
import pandas as pd
```

```
pd.DataFrame(distance)
```

```
step = 1000
data = []
for i in range(n):
  data.append(distance[i][step-1])

freq = {}
for i in data:
  try:
    freq[int(i)] += 1
  except KeyError:
    freq[int(i)] = 1
plt.scatter( list(freq.keys()), list(freq.values()) )
```
